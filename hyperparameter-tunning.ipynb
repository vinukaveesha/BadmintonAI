```python
import matplotlib.pyplot as plt
```

```python
# Epoch data for each sequence length
epochs = [1, 2, 3, 4, 5]

# SEQUENCE_LENGTH = 5
acc_5 = [0.4777, 0.7294, 0.8317, 0.8903, 0.9370]
val_acc_5 = [0.6933, 0.9867, 0.9867, 0.9867, 0.9867]
loss_5 = [1.3665, 1.1127, 0.9512, 0.8256, 0.7539]
val_loss_5 = [1.2001, 0.8939, 0.7868, 0.6920, 0.5913]

# SEQUENCE_LENGTH = 10
acc_10 = [0.3828, 0.6513, 0.6844, 0.7721, 0.8557]
val_acc_10 = [0.7067, 0.8800, 1.0000, 1.0000, 0.9867]
loss_10 = [1.4019, 1.0046, 0.9711, 0.9083, 0.8146]
val_loss_10 = [1.1068, 0.9496, 0.8774, 0.7971, 0.7120]

# SEQUENCE_LENGTH = 15
acc_15 = [0.3723, 0.5736, 0.5880, 0.6282, 0.6866]
val_acc_15 = [0.5467, 0.6800, 0.7733, 0.7733, 0.7733]
loss_15 = [1.4803, 1.1881, 1.0909, 1.0540, 0.9240]
val_loss_15 = [1.1865, 1.0768, 1.0085, 0.9507, 0.8951]

# SEQUENCE_LENGTH = 20
acc_20 = [0.4061, 0.6567, 0.7761, 0.8734, 0.9077]
val_acc_20 = [0.8667, 0.9467, 1.0000, 1.0000, 1.0000]
loss_20 = [1.3990, 1.1107, 0.9452, 0.8320, 0.7097]
val_loss_20 = [1.0398, 0.9786, 0.7546, 0.6616, 0.5823]
```

```python
# Plotting
plt.figure(figsize=(14, 10))

# Accuracy Plot
plt.subplot(2, 1, 1)
plt.plot(epochs, acc_5, marker='o', label='Train Acc (Seq 5)')
plt.plot(epochs, val_acc_5, marker='o', linestyle='--', label='Val Acc (Seq 5)')
plt.plot(epochs, acc_10, marker='s', label='Train Acc (Seq 10)')
plt.plot(epochs, val_acc_10, marker='s', linestyle='--', label='Val Acc (Seq 10)')
plt.plot(epochs, acc_15, marker='^', label='Train Acc (Seq 15)')
plt.plot(epochs, val_acc_15, marker='^', linestyle='--', label='Val Acc (Seq 15)')
plt.plot(epochs, acc_20, marker='d', label='Train Acc (Seq 20)')
plt.plot(epochs, val_acc_20, marker='d', linestyle='--', label='Val Acc (Seq 20)')
plt.title('Training and Validation Accuracy per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.xticks(epochs)
plt.ylim(0, 1.1)
plt.grid(True)
plt.legend()
```

```python
# Loss Plot
plt.subplot(2, 1, 2)
plt.plot(epochs, loss_5, marker='o', label='Train Loss (Seq 5)')
plt.plot(epochs, val_loss_5, marker='o', linestyle='--', label='Val Loss (Seq 5)')
plt.plot(epochs, loss_10, marker='s', label='Train Loss (Seq 10)')
plt.plot(epochs, val_loss_10, marker='s', linestyle='--', label='Val Loss (Seq 10)')
plt.plot(epochs, loss_15, marker='^', label='Train Loss (Seq 15)')
plt.plot(epochs, val_loss_15, marker='^', linestyle='--', label='Val Loss (Seq 15)')
plt.plot(epochs, loss_20, marker='d', label='Train Loss (Seq 20)')
plt.plot(epochs, val_loss_20, marker='d', linestyle='--', label='Val Loss (Seq 20)')
plt.title('Training and Validation Loss per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.xticks(epochs)
plt.ylim(0, 1.1)
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
